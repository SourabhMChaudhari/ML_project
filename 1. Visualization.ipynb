```python
import matplotlib.pyplot as plt

from matplotlib.colors import ListedColormap
```

```python
# plot USA state map 
geo_merge.plot(column='count', scheme="quantiles",  figsize=(25, 15),legend=True,cmap='Pastel1',linewidth=0.5, edgecolor='black')
plt.xlim(-130,-60)
plt.ylim(20,55)
# add state names and numbers 
for idx, row in geo_merge.iterrows():
    plt.text(row.geometry.centroid.x, row.geometry.centroid.y, "{}\n{}".format(row.NAME, row["count"]), size=10)

plt.title('Number of Reviews by States', fontsize=25)
plt.show()
```

```python
import datetime
import seaborn as sns
```

```python
# Convert the review_date column to datetime format
data['review_date'] = pd.to_datetime(data['review_date'])

# Create a new Month column
data['Month'] = data['review_date'].dt.month

# Group the data by state and Month and count the number of reviews
state_month_counts = data.groupby(['state', 'Month']).size().reset_index(name='count')

# Reshape the data into a pivot table
state_month_pivot = state_month_counts.pivot(index='state', columns='Month', values='count')

# Increase the size of the heatmap
plt.figure(figsize=(10,15))

# Create a heatmap of the pivot table
sns.heatmap(state_month_pivot, cmap='YlOrRd',linewidths=0.30,annot_kws={"fontsize":12},fmt='.0f')

plt.show()
```

```python
# Convert the review_date column to datetime format
data['review_date'] = pd.to_datetime(data['review_date'])

# Create a new Month column
data['Month'] = data['review_date'].dt.month

# Group the data by state and Month and count the number of reviews
state_month_counts = data.groupby(['region', 'Month']).size().reset_index(name='count')

# Reshape the data into a pivot table
state_month_pivot = state_month_counts.pivot(index='region', columns='Month', values='count')

# Increase the size of the heatmap
plt.figure(figsize=(10,6))

# Create a heatmap of the pivot table
sns.heatmap(state_month_pivot, cmap='RdYlGn',linewidths=0.30,annot_kws={"fontsize":12},fmt='.0f')

plt.show()
```

```python
usa_rate = data["star_rating"]
usa_rate_dataframe = usa_rate.value_counts().sort_index().reset_index()
usa_rate_dataframe = usa_rate_dataframe.rename(columns={'index': 'star_rating', 'star_rating': 'count'})


fig = px.pie(usa_rate_dataframe, values='count', names='star_rating', 
             title='Star Ratings in USA', 
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(hole=.4, hoverinfo="label+percent+name")

fig.update_layout(
    scene=dict(
        aspectratio=dict(x=1, y=1, z=0.7),
        camera_eye=dict(x=1.5, y=1.5, z=0.5)
    )
)

fig.show()
```

```python
import numpy as np
data = data.assign(sales=np.random.randint(1, 50, size=len(data)))
```

```python
#sales_by_region_month = df.groupby(['region', 'Month'])['sales'].sum().reset_index()
sales_by_region_month = data.groupby(['region', 'Month', 'product_category'])['sales'].sum().reset_index()

pivot_table = pd.pivot_table(sales_by_region_month, values='sales', index=['region', 'Month'], columns=['product_category'], fill_value=0)

pivot_table.plot(kind='bar', stacked=True, figsize=(10,8))
plt.xlabel('Region and Month')
plt.ylabel('Sales')
plt.title('Total Sales by Region and Month, by Product Category')
plt.legend(title='Product Category', bbox_to_anchor=(1.01, 1), loc='upper left')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# create data
region_sales = pd.pivot_table(data, values='sales', index=['region', 'Month'], aggfunc=np.sum)
region_sales = region_sales.reset_index()
regions = region_sales['region'].unique()

# create 3D plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

# create x, y, and z values
x = region_sales['region'].map({r:i for i, r in enumerate(regions)})
y = region_sales['Month']
z = region_sales['sales']

# create surface plot
ax.plot_trisurf(x, y, z, cmap='viridis')

# add labels
ax.set_xlabel('Region')
ax.set_xticks(np.arange(len(regions)))
ax.set_xticklabels(regions)
ax.set_ylabel('Month')
ax.set_zlabel('Total Sales')

plt.show()
```
